## Graph Class

```python
# helper class

class Graph:
    __access = 0  # total number of accessment to the graph data, which is used to evaluate your efficiency
    __access_norm = 10000

    __required_jobs = 3  # required number of different jobs in invitees

    __node_num = 0  # number of nodes (friends) in the graph
    __edge_num = 0  # number of edges (conflict relationships) in the graph
    __jobs = []  # __jobs[n] denotes the job of node n
    __job_num = 0  # number of different jobs in the graph
    __edge_list = []  # __edge_list[n][i] denotes the i-th neighbor of node n

    __invitees = []  # __invitees[n] == True: friend n is invited to the party

    def __init__(self, dataset: str):  # reading test case files and initialization
        with open(dataset, 'r') as f:
            nnum = int(f.readline()) #reading first line of file, which indicates number of friends ie nodes
            enum = 0

            self.__node_num = nnum

            job_line = f.readline() #reading 2nd line of file, jobs of each friend
            job_line = job_line.split(' ') #storing jobs into an array
            for i in range(nnum):
                self.__jobs.append(int(job_line[i])) #storing into list?

            for i in range(nnum):
                self.__edge_list.append([]) #list of n number of empty list, each list corr to each persons neighbours

            eline = f.readline() #to read the lines denoting conflicts
            while eline: #while there is a next line
                if ' ' in eline: 
                    eline = eline.split(' ')
                    node0 = int(eline[0]) #1st node to find on the graph
                    node1 = int(eline[1]) #2nd node to find on the graph
                    self.__edge_list[node0].append(node1) #appending enemy to own list from (line 30)
                    self.__edge_list[node1].append(node0) 
                    enum += 1 #counting num of edges i think
                eline = f.readline()

        for i in range(nnum):
            self.__edge_list[i].sort() #sort everybodys enemies in ascending order??
        self.__invitees = [False for _ in range(nnum)] #now set everyone's invited to party? status as False

        self.__edge_num = enum #number of edges
        self.__job_num = max(self.__jobs) + 1 #counting types of jobs (since start from index 0 need +1)

    def getNodeNum(self): #number of nodes
        return self.__node_num 

    def getEdgeNum(self): #number of edges aka number of pairs of enemies
        return self.__edge_num

    def getJobNum(self): #number of diff jobs
        return self.__job_num

    def getJobs(self, node: int): #job of node u looking for
        self.__access += 1
        return self.__jobs[node]

    def getDegree(self, node: int): #OH RETURNS NUMBER OF ENEMIES AHHHHH I SEEEEE no i lied huh why got repeats
        self.__access += 1
        return len(self.__edge_list[node])

    def getEdge(self, node: int, index: int): #um to access a particular enemy? but like in ascending order
        self.__access += 1
        return self.__edge_list[node][index]

    def getInvitee(self): #is it return who is alr invited
        return self.__invitees

    def setInvitee(self, node: int):  # invite friend represented by "node"
        self.__invitees[node] = True

    def setInviteeFromList(self, node_list): #huh
        for ni in node_list:
          # each element in node_list should be an integer in the range [0, node_num-1]
          if isinstance(ni, int) and ni >= 0 and ni < self.getNodeNum():
              self.setInvitee(ni)

    def isIndependentSet(self):  # check wether the current result is an independent set
        invitee_array = []
        for n in range(len(self.__invitees)):
            if self.__invitees[n]:
                invitee_array.append(n)

        for n in invitee_array:
            for nbr in self.__edge_list[n]:
                if self.__invitees[nbr]:
                    print("error: not a independent set") #means 0 marks
                    return False
        return True

    def enoughJobs(self): #check if enough diff jobs ah
        selected_jobs = set()
        for n in range(len(self.__invitees)):
            if self.__invitees[n]:
                selected_jobs.add(self.__jobs[n])
        if len(selected_jobs) >= self.__required_jobs:
            return True
        else:
            print("error: not enough jobs")
            return False

    def score(self):  # the function used for evaluation
        invitee_num = 0
        for n in range(len(self.__invitees)):
            if self.__invitees[n]:
                invitee_num += 1

        is_independent_set = self.isIndependentSet()
        enough_job = self.enoughJobs()
        print("Number of invitees: " + str(invitee_num) + "\n")
        print("Self access:" + str(self.__access) + "\n")
        grades = invitee_num - (self.__access / self.__access_norm)
        if is_independent_set and enough_job:
            return grades
        else:
            if is_independent_set:
              if self.getNodeNum() < 500:
                return max(grades - 2,0)
              else:
                return max(grades - 4,0)
            else:
              return 0
            

```

```python
def Invitees(g: Graph):
    '''
      return list "results" # including the list of node index recommended
      e.g., results = [0, 1] denotes that friends 0 and 1 will be invited
    '''
    results = []
    # do not change the code above

    # ******************** implement your code here ***********************
    
    totalFriends = g.getNodeNum()  

    candidate_list_num = {}
    for i in range(totalFriends):
        candidate_list_num[i] = g.getDegree(i)

    candidate_list_num = dict(sorted(candidate_list_num.items(), key=lambda item: item[1]))

    while len(list(candidate_list_num.keys())) > 0:
        results.append(list(candidate_list_num.keys())[0])

        appended = list(candidate_list_num.keys())[0]



        degAppended = candidate_list_num[appended]

        for i in range(degAppended):
            neighbor = g.getEdge(appended, i)

            if neighbor in candidate_list_num:
                del candidate_list_num[neighbor]

        del candidate_list_num[appended]
    
    return results


    
    
    
```

## Evaluation

We evaluate your submission by the function Graph::score(self), which computes the score using the following formula:

<ul>
<li> grades = invitee_num - (self.access / self.access_norm) - penalties_violate_job_constraint </li>
</ul>

The "invitee_num" is the number of friends you invited into the party. The self.__access is the number of accesses your submission have made to the data. We penalize the your submission based on self.__access to ensure the efficiency of your submission.

The "penalties_violate_job_constraint" has a value of either 2 or 4, depending on the number of friends ( n<500  or  n≥500 ) respectively. This is the number of points deducted if your proposed solution does not satisfy the constraint of having a minimum of 3 jobs in the guest list. Please note that if your solution result has any conflicts between 2 individuals, you will receive no points (grades = 0) for that test case.

In conclusion:

<ul>

<li> Any two of the invited friends cannot have conflict, i.e., no edge in the graph. </li>

otherwise, if the constraint can not be satisfied, please return an empty list. </li>
</ul>
